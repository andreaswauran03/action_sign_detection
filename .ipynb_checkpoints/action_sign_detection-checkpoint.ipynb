## MP Holistic and Detection Landmark

```python
import cv2
import mediapipe as mp
import numpy as np
import matplotlib.pyplot as plt
```

```python
mp_holistic = mp.solutions.holistic # Holistic Model
mp_drawing = mp.solutions.drawing_utils # Drawing Utilities
```

```python
def mediapipe_detection(image, model):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # BGR to RGB color
    image.flags.writeable = False                  # Image no longer writeable to avoid the problem of accidental changes in the image
    result = model.process(image)                  # Processing Image
    image.flags.writeable = True                   # Image is now writeable
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR) # RGB to BGR color
    return image, result
```

```python
def draw_style_landmark(image, results):
    # Drawing Face
    mp_drawing.draw_landmarks(image, results.face_landmarks, 
                              mp_holistic.FACEMESH_TESSELATION,
                              mp_drawing.DrawingSpec(color=(169, 255, 203), thickness=1, circle_radius=1),
                              mp_drawing.DrawingSpec(color=(222, 242, 144), thickness=1, circle_radius=1)) 
     # Drawing Right Hand
    mp_drawing.draw_landmarks(image, results.right_hand_landmarks, 
                              mp_holistic.HAND_CONNECTIONS,
                              mp_drawing.DrawingSpec(color=(145, 75, 255), thickness=2, circle_radius=4),
                              mp_drawing.DrawingSpec(color=(163, 2, 8), thickness=2, circle_radius=2))
     # Drawing Left Hand
    mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
                              mp_drawing.DrawingSpec(color=(49, 63, 216), thickness=2, circle_radius=4),
                              mp_drawing.DrawingSpec(color=(123, 26, 241), thickness=2, circle_radius=2))
    # Drawing Pose
    mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS,
                              mp_drawing.DrawingSpec(color=(49, 63, 216), thickness=2, circle_radius=4),
                              mp_drawing.DrawingSpec(color=(0, 79, 252), thickness=2, circle_radius=2))
```

```python
cap = cv2.VideoCapture(0) # capture video
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    while cap.isOpened():
        # Read Feed
        ret, frame = cap.read() # frame will contain the image frame, and ret will contain 
                                # a boolean value indicating whether reading the frame was successful.

        # Making Prediction
        image, results = mediapipe_detection(frame, holistic)

        # Drawing Landmarks
        draw_style_landmark(image, results)

        # Showing Image
        cv2.imshow("OpenCV Feed", image) 
        
        # Break cap video
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
        
    cap.release() #releases the resources used by the stamp object upon completion.
    cv2.destroyAllWindows() # closes all display windows that have been opened during program execution
```

```python
# Plotting Image Example
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

## Extracting Image to Array

```python
# Create Array
def get_array(results):
    array_pose = np.array([[res.x, res.y, res.z, res.visibility] for res in results.pose_landmarks.landmark]).flatten() if results.pose_landmarks else np.zeros(33*4)
    array_right_hand = np.array([[res.x, res.y, res.z] for res in results.right_hand_landmarks.landmark]).flatten() if results.right_hand_landmarks else np.zeros(21*3)
    array_left_hand = np.array([[res.x, res.y, res.z] for res in results.left_hand_landmarks.landmark]).flatten() if results.left_hand_landmarks else np.zeros(21*3)
    array_face = np.array([[res.x, res.y, res.z] for res in results.face_landmarks.landmark]).flatten() if results.face_landmarks else np.zeros(468*3)
    return np.concatenate([array_face, array_pose, array_left_hand, array_right_hand])
```

## Collecting Data

```python
import os
```

```python
DATA_PATH = "MP_Data"

try:
    os.makedirs(os.path.join(DATA_PATH))
except:
    pass
```

```python
actions = np.array(["paper", "rock", "scissor"])
```

```python
SEQUENCES = 30

SEQUENCES_LENGTH = 30

for action in actions:
    for seq in range(SEQUENCES):
        try:
            os.makedirs( os.path.join(DATA_PATH, action, str(seq)) )
        except:
            pass
```

```python
cap = cv2.VideoCapture(0) # capture video
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:

    # Loop through actions
    for action in actions:
        # Loop through videos
        for seq in range(SEQUENCES):
            # Loop through videos length
            for frame_num in range(SEQUENCES_LENGTH):
            
                # Read Feed
                ret, frame = cap.read() # frame will contain the image frame, and ret will contain 
                                        # a boolean value indicating whether reading the frame was successful.
                
                # Making Prediction
                image, results = mediapipe_detection(frame, holistic)

                # Drawing Landmarks
                draw_style_landmark(image, results)

                # Apply Text Collection
                if frame_num == 0:
                    cv2.putText(image, "Starting Collection", (120, 200),
                               cv2.FONT_HERSHEY_SCRIPT_COMPLEX, 1, (254, 0, 0), 3, cv2.LINE_AA)
                    cv2.putText(image, "Collect: {}, Number: {}".format(action, seq), (80, 50),
                               cv2.FONT_HERSHEY_DUPLEX, 1, (183, 19, 117), 1, cv2.LINE_AA)
                    cv2.waitKey(3000)
                else:
                    cv2.putText(image, "Collect: {}, Number: {}".format(action, seq), (80, 50),
                               cv2.FONT_HERSHEY_DUPLEX, 1, (183, 19, 117), 1, cv2.LINE_AA)
                # Extract Image to Array
                keypoint = get_array(results)
                # Get Path and save file
                npy_path = os.path.join(DATA_PATH, action, str(seq), str(frame_num))
                np.save(npy_path, keypoint)
                        
                # Showing Image
                cv2.imshow("OpenCV Feed", image) 
                
                # Break cap video
                if cv2.waitKey(10) & 0xFF == ord('q'):
                    break
    
    cap.release() #releases the resources used by the stamp object completion.
    cv2.destroyAllWindows() # closes all display windows that have been opened during program execution
```

### Preprocessiong

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
```

```python
# Creating Categorical Action number
label_map = {label:num for num, label in enumerate(actions)}
label_map
```

```python
# Taking a .npy file from dictionary
sequence, label = [], []
for action in actions:
    for seq in range(SEQUENCES):
        window = []
        for frame_num in range(SEQUENCES_LENGTH):
            arr = np.load(os.path.join(DATA_PATH, action, str(seq), "{}.npy".format(frame_num)))
            window.append(arr)
        sequence.append(window)
        label.append(label_map[action])
```

```python
# Create X dan y var
X = np.array(sequence)
y = to_categorical(label).astype(int)
```

```python
# Splitting dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

## Building and Train Models using LSTM

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import TensorBoard
```

```python
# Create a Log Folder
LOG_PATH = "logs"

try:
    os.makedirs(os.path.join(LOG_PATH))
except:
    pass
```

```python
# Create a tensorboard callback
tb_callback = TensorBoard(log_dir=os.path.join(LOG_PATH))
```

```python
# Building Architectural LSTM Neural Network
model = Sequential()
model.add(LSTM(32, activation="relu", return_sequences=True, input_shape=((X.shape)[1], (X.shape)[2])))
model.add(LSTM(64, activation="relu", return_sequences=False))
model.add(Dense(64, activation="relu"))
model.add(Dense(32, activation="relu"))
model.add(Dense(actions.shape[0], activation="softmax"))
```

```python
# Compile Model
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["categorical_accuracy"])
```

```python
# Train Model
model.fit(X_train, y_train, epochs=100, callbacks=[tb_callback])
```

```python
model.summary()
```

```python
# Making Prediction
y_pred = model.predict(X_test)
```
